### Line Plot

```python
## Importing the library

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
## Creating Random data for Line Plot
month = np.arange(1,13)


price = sorted(list(np.random.randint(200,400,12)))
print(month)
print(price)
```

```python
##  Plotting the data
plt.plot(month, price)

plt.title('Price trend of Products over the year')
plt.xlabel('Month')
plt.ylabel("Price")

plt.show()
```

```python
##  Plotting the data
plt.plot(month, price, color='g',linestyle='--', marker='*')

plt.title('Price trend of Products over the year')
plt.xlabel('Month', color='g')
plt.ylabel("Price", color='g')

plt.show()
```

```python
##  Plotting the data
#plt.plot(month, price, color='b', marker='*')
plt.plot(month, price,marker='.')
plt.title('Price trend over the year')
plt.xlabel('Month')
plt.ylabel("Price")

## To add a grid line to the plot
plt.grid(linestyle='--', linewidth='.2', color='b')
plt.show()
```

### Multiline plot

```python
## Creating some sample stock data

dt = pd.to_datetime(pd.date_range('2022-01-01','2022-01-20', periods=20))
google = np.random.randint(300,500,20)
yahoo = np.random.randint(200,400,20)
msft = np.random.randint(500,700,20)

df_stocks = pd.DataFrame({'Date':dt, 'Google':google, 'Yahoo':yahoo, 'Msft':msft})
df_stocks.head()

plt.figure(figsize=(10,8))
plt.plot(df_stocks.Date, df_stocks.Google, marker='*',label='Google')
plt.plot(df_stocks.Date, df_stocks.Yahoo, marker='*',label='Yahoo')
plt.plot(df_stocks.Date, df_stocks.Msft,marker='*',label='Microsoft')
plt.xlabel('Date',fontsize=15)
plt.ylabel('Price',fontsize=15)

plt.xticks(rotation=45,fontsize=12)
plt.title('Stock prices of Google, Yahoo and Microsoft')
plt.legend(bbox_to_anchor=[1,1],fontsize=12)
plt.show()
```

### Subplots

## Identifying Outliers in Claim

```python
qt25, qt75 = np.quantile(df_ins.claim, (.25, .75)).round(2)
print(qt25, qt75)
iqr = round(qt75 - qt25,2)
print(iqr)

upper_bound = round(qt75 + (1.5 * iqr),2)
print(upper_bound)

out_liers = df_ins[df_ins.claim > upper_bound]
```

#### Layered Pie Chart or nested Pie Chart

```python
fig, ax = plt.subplots()

size = 0.7
en_marks = np.random.randint(50,100,5)
sci_marks= np.random.randint(60,100,5)

cmap = plt.colormaps["tab20c"]
outer_colors = cmap(np.arange(5)*4)
inner_colors = cmap([1, 2, 5, 6, 9])

ax.pie(en_marks, radius=1, colors=outer_colors, labels=['A','B','C','D','E'], pctdistance=1.2, labeldistance=.8,
       autopct = lambda x: int(round(x/100*en_marks.sum(),0)), wedgeprops=dict(width=size, edgecolor='w'))

ax.pie(sci_marks, radius=1-size/2, colors=inner_colors,
       autopct = lambda x: int(round(x/100*sci_marks.sum(),0)), wedgeprops=dict(width=size, edgecolor='w'))

ax.set(aspect="equal", title='Pie plot with `ax.pie`')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy

labels = 'Frogs', 'Hogs', 'Dogs'
sizes = numpy.array([5860, 677, 3200])
colors = ['yellowgreen', 'gold', 'lightskyblue']

def absolute_value(val):
    a  = numpy.round(val/100.*sizes.sum(), 0)
    return a

plt.pie(sizes, labels=labels, colors=colors,
        autopct=absolute_value, shadow=True)

plt.axis('equal')
plt.show()
```
