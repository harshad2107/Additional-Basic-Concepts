### Hypothetical Testing Examples

**Question:**

The Violet industries produces furniture for home and office use. The production engineer claims that the average number of furniture items produced is 30. Test his claim.

Let, X is the number of furniture produced by the machine and the average number of furniture is 30

The Hypothesis is,
Null Hypothesis = Mean = 30
Alternate Hypothsis: Mean != 30

```python
## Importing the libraries
import pandas as pd
import numpy as np
```

```python
## Reading the file
df_production = pd.read_csv("E:\whatsapp Material\Statics and Probability\Production.csv")
df_production
```

```python
## Importing the libarary
from scipy.stats import ttest_1samp
```

```python
pop_mean = 30 
## Conducting the test
ttest_1samp(df_production["Items Produced"],pop_mean)
```

From the result of t test, we can see the p-value is .237, which is > 0.05.  Hence we accept the null hypothesis.  That ie. the average number of furniture produced by machine is 30.

__Question:__
The Blaze Industries Pvt. Ltd. claims that their manufactured electric bulbs have a life of 45000 hrs. Test the their claim at 5% level of significance.

**Solution:**

We can test the claim such the life of bulb is not equal to 45000.
However, it would be more meaningful to test for life of bulbs is less than 45000 hours since it would justify the claim is false.
If the life is more than 45000 hours the claim would be stronger. 
Thus we conduct a lower tail test
By default the in python the level of significance is 5%

The Null Hypothsis is average life of bulb is 45000.  i.e, Mean = 45000 Hrs
The Alternate hypthsis is Mean <= 45000 hrs


```python
## Reading the file
df_lifespan = pd.read_csv("E:\whatsapp Material\Statics and Probability\LifeSpan.csv")
df_lifespan
```

```python
## Conducting the test
t_test_less = ttest_1samp(df_lifespan['life'],45000)
t_test_less
```

```python
## To get the ttest_one_sided value, we devide the p_value by 2
p_value_less = t_test_less.pvalue/2
p_value_less
```

From the above result, we can see that the p_value is 0.022 which is less than 0.05.  Hence we reject the null hypothesis.  The claim of average life of light bulb is 45,000 hrs.  From the test we can conclude that the average life of a light bulb is less than 45,000 hours.

## Test for Normality

### Shapiro-Wilk Test

Testing the normality of samples which are less than 5000.


```python
## Import numpy library
import numpy as np
import random
```

```python
## Generating the sample data
rainfall = np.random.random(4500) * 100
print("Number of Samples generated:",len(rainfall))
print(random.sample(list(rainfall), 10))
rainfall[:10]
```

```python
## Importing the library
from scipy.stats import shapiro

## Conducting the test
stat, p = shapiro(rainfall)

print('The test statistic is ',stat,'and the p-value is ',p)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.kdeplot(rainfall)
plt.show()
```

From the result, we can see that the p-value is significantly less that 0.05.  Hence we are rejecting the null hypothesis.  We can conclude that the data is not normally distributed.

### Jarque-Bera Test

To test the normality of sample size more than 5000.


```python
## Generating the data
diam = np.random.normal(loc=4, scale=0.7, size=6000) + np.random.normal(loc=5, scale=0.1, size=6000)
print('The Number of samples generated : ',len(diam))

print()
print("Some sample data")
random.sample(list(diam),20)

```

```python
## Importing the library
from scipy.stats import jarque_bera

## Conducting the test
stat, p = jarque_bera(diam)

print('The test statistic is ',stat,'and the p-value is ',p)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#sns.displot(diam, kde=True)
sns.kdeplot(diam)
plt.show()
```

From the above result, we can see that the p-value is less than 0.05.  Hence we accept the null hypothesis.  We can conclude that the data is normally distributed.
